```python
data['rec_dec'] = mba263.ntile(data['sincepurch'],10)
data['freq_dec'] = mba263.ntile(data['numords'],10)
data['mv_dec'] = mba263.ntile(data['dollars'],10)
data['freq_dec'] = 9 - data['freq_dec']
data['mv_dec'] = 9 - data['mv_dec']
data['rfmindex_idex'] = 100 * data['rec_dec'] + 10 * data['freq_dec'] + data['mv_dec'] 
```

```python
data['response_rate']=mba263.get_means(data['res1'],data['rfmindex_idex'])
```

```python
data['mailto_iq']=(data['response_rate']>0.0235)*1
```
